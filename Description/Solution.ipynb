# Sigmoid

```python
import math

x = 2
y = 1 / (1 + math.exp(-x))
print(y)
```

# Tanh

```python
import math

x = 2
y = 2/(1 + math.exp(-2*x)) - 1
print(y)
```

# Softplus

```python
import math

x = 2
y = math.log(1 + math.exp(x))
print(y)
```

# ReLU

```python
x = -2.0
if x<=0:
    y = 0.0
else:
    y = x
print(y)
```

# LeakyReLU

```python
x = -1.2
if x<=0:
    y = x*0.01
else:
    y = x
print(y)
```

# ELU

```python
import math

alpha = 0.1
x = -4
if x<=0:
    y = alpha*(math.e**x-1)
else:
    y = x
print(y)
```

# PReLU

```python
alpha = 0.1
x = -4
if x<=0:
    y = x*alpha
else:
    y = x
print(y)
```

# Swish

```python
import math

beta = 1.0
x = -4
y = x * (1 / (1 + math.e**(-beta*x)))
print(y)
```
